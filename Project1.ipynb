## Project Introduction
This project will investigate into which groups are hit the hardest by inflation in the United States. We will evaluate and compare inflation trends with wage changes across different income levels and industries to understand how inflation affects different parts of the population.
## Research Questions:
1. Does inflation affect low-income groups more severely than high-income ones?
2. Do wages keep up with inflation across industries?
3. Are certain geographical regions hit harder by inflation than others?
