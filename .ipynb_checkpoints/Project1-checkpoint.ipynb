Project introduction : 
Initial description of the project topic : This project will dive into who/ what groups are hit the hardest by inflation. We will evaluate and compare inflation trends with wage changes across different industries.
Potential research questions you plan to address:
- Does inflation affect low-income groups more severely than high-income ones?
- Do wages keep up with inflation across industries?
